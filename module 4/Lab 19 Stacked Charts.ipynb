### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 65
}

# Replace the Age column with the numerical values
df['Age_num'] = df['Age'].map(age_mapping)
# Group by Age and calculate median values
age_group_satisfaction = df.groupby('Age_num')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot the stacked chart
age_group_satisfaction.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#1f77b4', '#ff7f0e'])
plt.title('Median Job Satisfaction Points Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Median Satisfaction Points')
plt.legend(['JobSatPoints_6', 'JobSatPoints_7'], loc='upper right')
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
# Group employment type by job satisfaction
employment_by_satisfaction = df.groupby(['JobSatPoints_6', 'Employment']).size().unstack(fill_value=0)

# Plot the stacked chart
employment_by_satisfaction.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title('Employment Types by Job Satisfaction Levels')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Count')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here
# Filter data for age group 30-35
age_30_35 = df[(df['Age_num'] >= 30) & (df['Age_num'] <= 35)]

# Group and calculate median values
comp_satisfaction_30_35 = age_30_35[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# Plot the stacked chart
comp_satisfaction_30_35.plot(kind='bar', stacked=True, figsize=(8, 6), color=['#2ca02c', '#d62728'])
plt.title('Compensation and Job Satisfaction (Age 30-35)')
plt.ylabel('Median Values')
plt.xticks(ticks=[], labels=[])
plt.legend(['Compensation', 'JobSatPoints_6'])
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
# Group by Age and calculate median values
compensation_satisfaction_by_age = df.groupby('Age_num')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# Plot the stacked chart
compensation_satisfaction_by_age.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#2ca02c', '#d62728'])
plt.title('Median Compensation and Job Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Values')
plt.legend(['Compensation', 'JobSatPoints_6'], loc='upper right')
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
# Expand database preferences by age group
database_preferences = df['DatabaseWantToWorkWith'].str.split(';').explode()
db_by_age = database_preferences.groupby([df['Age_num'], database_preferences]).size().unstack(fill_value=0).iloc[:, :5]

# Plot the stacked chart
db_by_age.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Preferred Databases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Databases', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
# Group employment type by job satisfaction
employment_by_satisfaction = df.groupby(['JobSatPoints_6', 'Employment']).size().unstack(fill_value=0)

# Plot the stacked chart
employment_by_satisfaction.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title('Employment Types by Job Satisfaction Levels')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Count')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
# Expand language preferences by age group
languages_preferences = df['LanguageAdmired'].str.split(';').explode()
lang_by_age = languages_preferences.groupby([df['Age_num'], languages_preferences]).size().unstack(fill_value=0).iloc[:, :5]

# Plot the stacked chart
lang_by_age.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='plasma')
plt.title('Preferred Programming Languages by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Languages', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
##Write your code here
# Expand platform preferences by employment type
platform_preferences = df['PlatformAdmired'].str.split(';').explode()
platform_by_employment = platform_preferences.groupby([df['Employment'], platform_preferences]).size().unstack(fill_value=0).iloc[:, :5]

# Plot the stacked chart
platform_by_employment.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='spring')
plt.title('Technology Adoption by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.legend(title='Platforms', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
