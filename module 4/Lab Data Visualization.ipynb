**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
QUERY = "SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL AND CompTotal < 1000000;"
data = pd.read_sql_query(QUERY, conn)

plt.hist(data['CompTotal'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Compensation')
plt.xlabel('Compensation (CompTotal)')
plt.ylabel('Frequency')
plt.show()

```

**Box Plots**

Plot a box plot of Age.


```python
QUERY = "SELECT Age FROM main WHERE Age IS NOT NULL;"
data = pd.read_sql_query(QUERY, conn)

# Plot box plot
sns.boxplot(y=data['Age'], color='orange')
plt.title('Box Plot of Age')
plt.xlabel('Age Groups')
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = "SELECT Age, WorkExp FROM main WHERE Age IS NOT NULL AND WorkExp IS NOT NULL;"
data = pd.read_sql_query(QUERY, conn)

# Scatter plot
plt.scatter(data['WorkExp'], data['Age'], alpha=0.5, color='green')
plt.title('Scatter Plot of Age vs Work Experience')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Age')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
query = """
SELECT 
    TimeSearching, 
    Frustration, 
    CASE 
        WHEN Age = 'Under 18 years old' THEN 15
        WHEN Age = '18-24 years old' THEN 21
        WHEN Age = '25-34 years old' THEN 30
        WHEN Age = '35-44 years old' THEN 40
        WHEN Age = '45-54 years old' THEN 50
        WHEN Age = '55-64 years old' THEN 60
        WHEN Age = '65 years or older' THEN 70
        ELSE NULL
    END AS AgeNumeric
FROM main
WHERE TimeSearching IS NOT NULL AND Frustration IS NOT NULL AND Age IS NOT NULL;
"""

# Load data into DataFrame
data = pd.read_sql_query(query, conn)
import numpy as np

# Add jitter to avoid overlaps
jitter_time = np.random.uniform(-0.5, 0.5, size=len(data))
jitter_frustration = np.random.uniform(-0.5, 0.5, size=len(data))
# Bubble plot
plt.scatter(
    data['TimeSearching'], 
    data['Frustration'], 
    s=data['AgeNumeric'] * 10,  # Bubble size based on numeric Age
    alpha=0.5, 
    color='purple'
)
plt.title('Bubble Plot: TimeSearching vs Frustration')
plt.xlabel('Time Spent Searching')
plt.ylabel('Frustration Level')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
QUERY = """
SELECT DatabaseWantToWorkWith, COUNT(*) as count
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
GROUP BY DatabaseWantToWorkWith
ORDER BY count DESC
LIMIT 5;
"""
data = pd.read_sql_query(QUERY, conn)

# Pie chart
plt.pie(data['count'], labels=data['DatabaseWantToWorkWith'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Top 5 Databases Respondents Want to Work With')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
query = """
SELECT 
    CASE 
        WHEN Age = 'Under 18 years old' THEN 17
        WHEN Age = '18-24 years old' THEN 21
        WHEN Age = '25-34 years old' THEN 30
        WHEN Age = '35-44 years old' THEN 40
        WHEN Age = '45-54 years old' THEN 50
        WHEN Age = '55-64 years old' THEN 60
        WHEN Age = '65 years or older' THEN 70
    END AS AgeNumeric,
    CASE 
        WHEN TimeSearching = 'Less than 15 minutes a day' THEN 7.5
        WHEN TimeSearching = '15-30 minutes a day' THEN 22.5
        WHEN TimeSearching = '30-60 minutes a day' THEN 45
        WHEN TimeSearching = '60-120 minutes a day' THEN 90
        WHEN TimeSearching = 'Over 120 minutes a day' THEN 150
        ELSE NULL
    END AS TimeSearchingNumeric,
    CASE 
        WHEN TimeAnswering = 'Less than 15 minutes a day' THEN 7.5
        WHEN TimeAnswering = '15-30 minutes a day' THEN 22.5
        WHEN TimeAnswering = '30-60 minutes a day' THEN 45
        WHEN TimeAnswering = '60-120 minutes a day' THEN 90
        WHEN TimeAnswering = 'Over 120 minutes a day' THEN 150
        ELSE NULL
    END AS TimeAnsweringNumeric
FROM main
WHERE Age IS NOT NULL 
  AND TimeSearching IS NOT NULL 
  AND TimeAnswering IS NOT NULL;
"""

data = pd.read_sql_query(query, conn)

# Calculate medians
median_searching = data.groupby('AgeNumeric')['TimeSearchingNumeric'].median()
median_answering = data.groupby('AgeNumeric')['TimeAnsweringNumeric'].median()

# Stacked bar chart
ages = median_searching.index
plt.bar(ages, median_searching, label='Median Time Searching')
plt.bar(ages, median_answering, bottom=median_searching, label='Median Time Answering')
plt.title('Stacked Bar Chart: TimeSearching and TimeAnswering')
plt.xlabel('Age (Numeric)')
plt.ylabel('Time (Minutes)')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY = """
SELECT Age, CompTotal 
FROM main 
WHERE Age BETWEEN '45' AND '60' AND CompTotal IS NOT NULL;
"""
data = pd.read_sql_query(QUERY, conn)

median_comp = data.groupby('Age')['CompTotal'].median()

# Line chart
plt.plot(median_comp.index, median_comp.values, marker='o', color='blue')
plt.title('Median Compensation by Age (45-60)')
plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM main
GROUP BY MainBranch
ORDER BY count DESC;
"""
data = pd.read_sql_query(QUERY, conn)

# Horizontal bar chart
plt.barh(data['MainBranch'], data['count'], color='teal')
plt.title('Horizontal Bar Chart: MainBranch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```
