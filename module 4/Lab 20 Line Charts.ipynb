### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 65
}

# Replace the Age column with the numerical values
df['Age_num'] = df['Age'].map(age_mapping)
# Group by Age and calculate median compensation
median_comp_by_age = df.groupby('Age_num')['ConvertedCompYearly'].median()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age, marker='o', color='b')
plt.title('Median Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.grid(True)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
# Filter data for age range 25 to 45 and calculate median compensation
filtered_comp_by_age = median_comp_by_age[(median_comp_by_age.index >= 25) & (median_comp_by_age.index <= 45)]

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(filtered_comp_by_age, marker='o', color='g')
plt.title('Median Compensation for Ages 25 to 45')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.grid(True)
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# Substitute with the available experience-related column (e.g., 'WorkExp')
experience_satisfaction = df.groupby('WorkExp')['JobSatPoints_6'].median()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(experience_satisfaction, marker='o', color='r')
plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Experience Level (Years)')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
# Group by experience level and calculate median compensation
compensation_by_experience = df.groupby('WorkExp')['ConvertedCompYearly'].median()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(compensation_by_experience, marker='o', color='purple')
plt.title('Median Compensation by Experience Level')
plt.xlabel('Experience Level (Years)')
plt.ylabel('Median Yearly Compensation')
plt.grid(True)
plt.show()
```
